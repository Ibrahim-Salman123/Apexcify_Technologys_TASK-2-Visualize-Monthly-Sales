## <span style=color:yellow;>****_Convert ORDER DATE to datetime_****</span>

```python
df["ORDER DATE"] = pd.to_datetime(df["ORDER DATE"], format="%d/%m/%Y")
df
```

## <span style=color:yellow;>****_Extract Month-Year_****</span>

```python
df["Month"] = df["ORDER DATE"].dt.to_period("M")
df
```

## <span style=color:yellow;>****_Aggregate Sales by Month_****</span>

```python
monthly_sales = df.groupby("Month")["SALES"].sum().reset_index()
```

## <span style=color:yellow;>****_Plotting a Line Chart showing the Sales Trend over the Year_****</span>

```python
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales["Month"].astype(str), monthly_sales["SALES"], marker="o", linestyle="-")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Monthly Sales Trend")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
